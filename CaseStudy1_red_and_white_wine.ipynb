```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

df_red = pd.read_csv('winequality-red.csv',sep=";")
df_white = pd.read_csv('winequality-white.csv',sep=";")
df_white.head(3)
```

```python
df_white.shape # 4898
df_red.shape # 1599
```

```python
# create color array for red and white datafram and repeat with "white" or "red" no. of rows-times
color_red = np.repeat("red",1599)
color_white = np.repeat("white",4898)
```

```python
# add respective arrays to red and white dataframes
df_red['color'] = color_red
df_white['color'] = color_white
```

```python
# check for this
df_red.head(3)
```

```python
# check for this
df_white.head(3)
```

```python
# combine dataframes
wine_df = df_white.append(df_red,ignore_index=True)
wine_df.head(10)
```

```python
# save it as winemerged
wine_df.to_csv('winemerged.csv',index=False)
```
