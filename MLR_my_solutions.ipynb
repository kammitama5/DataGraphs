```python
import numpy as np
import pandas as pd
import statsmodels.api as sm;

df = pd.read_csv("house_prices.csv")
df.head()
```

```python
df['intercept'] = 1
lm = sm.OLS(df['price'], df[['intercept','bathrooms','bedrooms','area']]) # lienar model,
# plot price against all the other quantitative values
results = lm.fit()
results.summary()
```

```python
lm1 = sm.OLS(df['price'], df[['intercept','bathrooms']]) # lienar model,
# plot price against all the other quantitative values
results1 = lm1.fit()
results1.summary()
```

```python
lm2 = sm.OLS(df['price'], df[['intercept','bedrooms']]) # lienar model,
# plot price against all the other quantitative values
results2 = lm2.fit()
results2.summary()
```

```python
lm3 = sm.OLS(df['price'], df[['intercept','area']]) # lienar model,
# plot price against all the other quantitative values
results3 = lm3.fit()
results3.summary()
```

```python
lm5 = sm.OLS(df['price'],df[['intercept','bathrooms','bedrooms','area']])
results5 = lm5.fit()
results5.summary()
```

```python
# transpose, inverse, transpose, dot product 
x = df[['intercept', 'bathrooms','bedrooms','area']]
y = df['price']
```

```python
# chain operations together
np.dot(np.dot(np.linalg.inv(np.dot(x.transpose(),x)),x.transpose()),y)
```
