```python
import pandas as pd
import numpy as np

df = pd.read_csv('cancer.csv')
df.head()
```

```python
df.columns = ['id', 'diagnosis','radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_SE', 'texture_SE', 'perimeter_SE', 'area_SE', 'smoothness_SE','compactness_SE', 'concavity_SE', 'concave_points_SE', 'symmetry_SE', 'fractal_dimension_SE', 'radius_max', 'texture_max', 'perimeter_max', 'area_max', 'smoothness_max', 'compactness_max', 'concavity_max', 'concave_points_max', 'symmetry_max', 'fractal_dimension_max']
```

```python
df.head(3)
```

```python
# replace all non-null values with mean value
mean = df.mean()
df.fillna(mean, inplace=True)
# confirm the correction with info()
df.info()
df.head(3)
```

```python
# check for duplicates
sum(df.duplicated()) # 0
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
# shouldn't do anything
```

```python
# confirm correction by rechecking for duplicates in the data
sum(df.duplicated())
```

```python
# remove "_mean" from column names
new_labels = []
for col in df.columns:
    if '_mean' in col:
        new_labels.append(col[:-5]) # append without "_mean", which is last five chars
    else:
        new_labels.append(col)
new_labels # no column name has "_mean" in it
```

```python
# assign new labels to columns in dataframe
df.columns = new_labels
# check that changes have taken place
df.head(3)
```

```python
# save this for later
df.to_csv('cancer_data_edited.csv')
```
