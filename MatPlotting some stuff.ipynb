```python
# import # Load `winequality_edited.csv`
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

df = pd.read_csv('winemerged.csv')
df.head(3)
```

```python
# use query to select each group and get its mean quality
median = df['alcohol'].median()
low = df.query('alcohol < {}'.format(median))
high = df.query('alcohol >= {}'.format(median))

mean_quality_low = low['quality'].mean()
mean_quality_high = high['quality'].mean()
```

```python
# Create a bar chart with proper labels
locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating');
```

```python
# Use query to select each group and get its mean quality
df = df.rename(columns={'residual sugar': 'residual_sugar'})
df.head(3)
```

```python
# create a bar chart one with low residual sugar and one bar for high residual sugar wine samples
mean_residual_sugar = df['residual_sugar'].mean()
mean_residual_sugar
```

```python
# Use query to select each group and get its mean quality
median_residual_sugar = df['residual_sugar'].median()
low_residual_sugar = df.query('residual_sugar < {}'.format(median_residual_sugar))
high_residual_sugar = df.query('residual_sugar >= {}'.format(median_residual_sugar))

mean_low_residual_sugar = low_residual_sugar['quality'].mean()
mean_high_residual_sugar = high_residual_sugar['quality'].mean()
```

```python
# Create a bar chart with proper labels
locations = [1, 2]
heights = [mean_residual_sugar, mean_high_residual_sugar]
labels = ['low_residual_sugar', 'high_residual_sugar']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Residual Sugar')
plt.xlabel('Residual Sugar')
plt.ylabel('Average Quality Rating');
```
